```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image

# Path ke folder utama yang berisi subfolder gambar pepaya
folder_pepaya = '/content/drive/MyDrive/BISMILLAH TA/pepaya'  # Ganti dengan path kamu

# Dapatkan daftar semua subfolder di folder pepaya
subfolders = [f.path for f in os.scandir(folder_pepaya) if f.is_dir()]

# Buat figure dan axes untuk menampilkan gambar secara horizontal
fig, axes = plt.subplots(1, 3, figsize=(10, 15))
fig.suptitle("Gambar pepaya - Zoom 1.5x", fontsize=16)

# Faktor zoom
zoom_factor = 1.5

for i, ax in enumerate(axes.flat):
    pepaya_name = os.path.basename(subfolders[i])
    image_files = [f for f in os.listdir(subfolders[i]) if f.endswith(('.jpg', '.png', '.jpeg'))]

    # Pilih satu gambar secara acak
    if image_files:
        selected_image = random.choice(image_files)
        img_path = os.path.join(subfolders[i], selected_image)

        # Baca gambar
        img = Image.open(img_path)
        img_array = mpimg.pil_to_array(img)

        # Hitung ukuran gambar
        height, width, _ = img_array.shape
        new_height = int(height / zoom_factor)
        new_width = int(width / zoom_factor)

        # Hitung koordinat cropping (dari tengah)
        x_start = (width - new_width) // 2
        x_end = x_start + new_width
        y_start = (height - new_height) // 2
        y_end = y_start + new_height

        # Tampilkan gambar hasil zoom
        ax.imshow(img_array[y_start:y_end, x_start:x_end])
        ax.axis('off')
        ax.set_title(pepaya_name)

plt.tight_layout()
plt.show()
```

```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image, ImageEnhance
import numpy as np
import cv2

# Path ke folder dataset pepaya
folder_pepaya = '/content/drive/MyDrive/BISMILLAH TA/pepaya'

# Ambil semua subfolder
subfolders = [f.path for f in os.scandir(folder_pepaya) if f.is_dir()]

# Konfigurasi brightness & contrast
brightness_factor = 2.0
contrast_factor = 2.0

# Figure untuk grayscale
fig, axes = plt.subplots(1, 3, figsize=(10, 15))
fig.suptitle("Gambar Pepaya - Grayscale setelah Brightness & Contrast", fontsize=14)

grayscale_images = []  # Simpan hasil grayscale untuk tahap berikutnya (Otsu)

for i, ax in enumerate(axes.flat):
    if i < len(subfolders):
        folder = subfolders[i]
        category = os.path.basename(folder)
        image_files = [f for f in os.listdir(folder) if f.endswith(('.jpg', '.png', '.jpeg'))]

        if image_files:
            selected_image = random.choice(image_files)
            img_path = os.path.join(folder, selected_image)

            # Open & enhance
            img = Image.open(img_path)
            img = ImageEnhance.Brightness(img).enhance(brightness_factor)
            img = ImageEnhance.Contrast(img).enhance(contrast_factor)

            # Convert ke numpy array (RGB), lalu ke grayscale (OpenCV butuh BGR)
            img_np = np.array(img)
            img_bgr = cv2.cvtColor(img_np, cv2.COLOR_RGB2BGR)
            gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

            # Simpan untuk tahap Otsu
            grayscale_images.append((gray, category))

            ax.imshow(gray, cmap='gray')
            ax.axis('off')
            ax.set_title(category)
    else:
        ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Figure untuk Otsu
fig, axes = plt.subplots(1, 3, figsize=(10, 15))
fig.suptitle("Segmentasi Otsu dari Gambar Grayscale", fontsize=14)

for i, (gray_img, label) in enumerate(grayscale_images):
    if i < len(axes.flat):
        ax = axes.flat[i]

        # Blur untuk stabilitas Otsu
        blur = cv2.GaussianBlur(gray_img, (5, 5), 0)

        # Otsu Thresholding
        _, otsu_result = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

        ax.imshow(otsu_result, cmap='gray')
        ax.axis('off')
        ax.set_title(label)
    else:
        break

plt.tight_layout()
plt.show()
```

```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from skimage.feature import graycomatrix, graycoprops
from PIL import Image, ImageEnhance

folder_pepaya = 'pepaya'
```

```python
def resize_image(img, size=(128, 128)):
  return img.resize(size)

def rotation_image(img, angle=35):
  return img.rotate(angle)

def flip_image(img, type="horizontal"):
  if type == "horizontal":
    image = img.transpose(Image.FLIP_LEFT_RIGHT)
  elif type == "vertical":
    image = img.transpose(Image.FLIP_TOP_BOTTOM)

  return image

def brightness_image(img, brightness=1.2, contrast=1.5):
    # Penyesuaian kecerahan
    enhancer_brightness = ImageEnhance.Brightness(img)
    img_brightness = enhancer_brightness.enhance(brightness)

    # Penyesuaian kontras
    enhancer_contrast = ImageEnhance.Contrast(img_brightness)
    img_contrast = enhancer_contrast.enhance(contrast)

    return img_contrast

def glcm_feature(image):
    # if image is PIL convert to np.array
    img = np.array(image)

    glcm = graycomatrix(img, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)

    # Extract features
    contrast = graycoprops(glcm, 'contrast')[0, 0]
    dissimilarity = graycoprops(glcm, 'dissimilarity')[0, 0]
    homogeneity = graycoprops(glcm, 'homogeneity')[0, 0]
    energy = graycoprops(glcm, 'energy')[0, 0]
    correlation = graycoprops(glcm, 'correlation')[0, 0]
    asm = graycoprops(glcm, 'ASM')[0, 0]

    return [contrast, dissimilarity, homogeneity, energy, correlation, asm]
```

```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
from skimage.feature import graycomatrix, graycoprops
from PIL import Image

# Tentukan path ke folder dataset pepaya
folder_pepaya = '/content/drive/MyDrive/BISMILLAH TA/pepaya'  # Ganti dengan path yang sesuai

# Periksa apakah folder dataset tersedia
if not os.path.exists(folder_pepaya):
    raise FileNotFoundError(f"Folder dataset tidak ditemukan: {folder_pepaya}")

# Dapatkan daftar semua subfolder di dalam folder pepaya
subfolders = [f.path for f in os.scandir(folder_pepaya) if f.is_dir()]

# Periksa apakah ada subfolder
if not subfolders:
    raise ValueError(f"Tidak ada subfolder dalam folder dataset: {folder_pepaya}")

features = []
labels = []

# Fungsi augmentasi
def rotation_image(image):
    return image.rotate(random.choice([90, 180, 270]))

def flip_image(image):
    return image.transpose(Image.FLIP_LEFT_RIGHT)

def brightness_image(image):
    enhancer = ImageEnhance.Brightness(image)
    return enhancer.enhance(random.uniform(0.8, 1.2))

def resize_image(image, size=(128, 128)):
    return image.resize(size)

def glcm_feature(image):
    gray_image = np.array(image)
    distances = [1]
    angles = [0]
    glcm = graycomatrix(gray_image, distances, angles, 256, symmetric=True, normed=True)
    contrast = graycoprops(glcm, 'contrast')[0, 0]
    homogeneity = graycoprops(glcm, 'homogeneity')[0, 0]
    energy = graycoprops(glcm, 'energy')[0, 0]
    correlation = graycoprops(glcm, 'correlation')[0, 0]
    return [contrast, homogeneity, energy, correlation]

augmentation = [rotation_image, flip_image, brightness_image]

# Loop melalui setiap subfolder
for label in os.listdir(folder_pepaya):
    label_path = os.path.join(folder_pepaya, label)

    if os.path.isdir(label_path):
        for image_file in os.listdir(label_path):
            image_path = os.path.join(label_path, image_file)

            try:
                image = Image.open(image_path).convert('L')  # Konversi ke grayscale
                image = resize_image(image)

                # Ekstrak fitur GLCM
                image_features = glcm_feature(image)
                features.append(image_features)
                labels.append(label)

                # Augmentasi gambar
                for aug in augmentation:
                    aug_image = aug(image)
                    aug_features = glcm_feature(aug_image)
                    features.append(aug_features)
                    labels.append(label)

            except Exception as e:
                print(f"Error loading image {image_path}: {e}")

print(f"Total data yang berhasil diproses: {len(features)}")
```

ALGORITMA KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Rentang nilai K ganjil dari 1 sampai 9
k_values = range(1, 10, 2)

# Menyimpan hasil akurasi
results = []

for k in k_values:
    # Inisialisasi dan latih model
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train, y_train)

    # Prediksi pada data uji
    y_pred = knn_model.predict(X_test)

    # Hitung akurasi
    accuracy = accuracy_score(y_test, y_pred)

    # Simpan hasil
    results.append({'K': k, 'Akurasi': accuracy})

# Tampilkan hasil dalam DataFrame
import pandas as pd
results_df = pd.DataFrame(results)
print(results_df)

# Menemukan nilai K terbaik berdasarkan akurasi tertinggi
optimal_k_test = results_df.loc[results_df['Akurasi'].idxmax(), 'K']
print(f"Nilai K terbaik berdasarkan akurasi di data uji: {optimal_k_test}")
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Implementasi model knn
# Menggunakan nilai K terbaik dari data uji
knn_model = KNeighborsClassifier(n_neighbors=optimal_k_test) # Mengganti optimal_k_accuracy dengan optimal_k_test
knn_model.fit(X_train, y_train)
```

```python
# Prediksi hasil KNN
y_pred_knn = knn_model.predict(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Implementasi model knn
# Menggunakan nilai K terbaik dari data uji
knn_model = KNeighborsClassifier(n_neighbors=optimal_k_test) # Mengganti optimal_k_accuracy dengan optimal_k_test
knn_model.fit(X_train, y_train)

# Prediksi hasil KNN
y_pred_knn = knn_model.predict(X_test)

# Evaluasi KNN
conf_matrix_knn = confusion_matrix(y_test, y_pred_knn)
print("Confusion Matrix - KNN:\n", conf_matrix_knn)
print("Classification Report - KNN:\n", classification_report(y_test, y_pred_knn))
print("Accuracy Score - KNN:", accuracy_score(y_test, y_pred_knn))
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import numpy as np # Make sure numpy is imported for np.unique

# Visualisasi Confusion Matrix
def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y), yticklabels=np.unique(y))
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(title)
    plt.show()

plot_confusion_matrix(y_test, y_pred_knn, "Confusion Matrix - KNN")
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression

# Inisialisasi model Logistic Regression
logreg_model = LogisticRegression(random_state=42, max_iter=1000, C=1.0, solver='lbfgs')

# Melatih model dengan data pelatihan
logreg_model.fit(X_train, y_train)
```

```python
# Prediksi hasil Logistic Regression
y_pred_lr = logreg_model.predict(X_test)
```

```python
# Evaluasi Logistic Regression
# Mengganti y_pred_nb dengan y_pred_lr untuk mengevaluasi model Logistic Regression
conf_matrix_lr = confusion_matrix(y_test, y_pred_lr)
print("Confusion Matrix - Logistic Regression:\n", conf_matrix_lr)
print("Classification Report - Logistic Regression:\n", classification_report(y_test, y_pred_lr))
print("Accuracy Score - Logistic Regression:", accuracy_score(y_test, y_pred_lr))
```

```python
# Visualisasi Confusion Matrix
def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y), yticklabels=np.unique(y))
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(title)
    plt.show()
# Mengganti y_pred_nb dengan y_pred_lr dan memperbarui judul plot
plot_confusion_matrix(y_test, y_pred_lr, "Confusion Matrix - Logistic Regression")
```

```python
# Akurasi dari kedua model (ganti dengan nilai akurasi Anda)
knn_accuracy = accuracy_score(y_test, y_pred_knn)
lr_accuracy = accuracy_score(y_test, y_pred_lr)

print(f"Accuracy KNN: {knn_accuracy:.4f}")
print(f"Accuracy Logistic Regression: {lr_accuracy:.4f}")
```

```python
# Data untuk diagram batang
# Mengganti 'Naive Bayes' dengan 'Logistic Regression' dan nb_accuracy dengan lr_accuracy
models = ['K-Nearest Neighbor', 'Logistic Regression']
accuracies = [knn_accuracy, lr_accuracy]
# Membuat diagram batang
plt.figure(figsize=(8, 6))
bars = plt.bar(models, accuracies, color=['blue', 'orange'])

# Menambahkan label nilai akurasi di atas setiap batang
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, round(yval, 4), ha='center', va='bottom')

plt.title('Perbandingan Akurasi Model')
plt.xlabel('Model')
plt.ylabel('Akurasi')
plt.ylim([0, 1.1])  # Mengatur batas sumbu y agar lebih jelas
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
